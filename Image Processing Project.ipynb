```python
from tkinter import *
import cv2
from PIL import ImageTk, Image
from tkinter import filedialog
from skimage import filters, img_as_float
import os
import matplotlib.pyplot as plt
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
import numpy as np
from skimage.color import rgb2gray, gray2rgb
from skimage.data import gravel
from skimage.filters import gaussian, difference_of_gaussians, window
from scipy.fftpack import fftn, fftshift
from skimage.segmentation import active_contour
from skimage import exposure
from skimage.exposure import match_histograms, equalize_hist,rescale_intensity
from matplotlib.pyplot import imread, imshow, show, subplot, title, get_cmap, hist
from skimage import data, color
from skimage.transform import rescale, resize, downscale_local_mean, rotate, swirl
from tkinter import simpledialog
from skimage.util import crop
from scipy.ndimage import map_coordinates
from skimage.morphology import convex_hull_image
from skimage import data, img_as_float
from skimage.util import invert
from skimage.morphology import disk
from skimage.filters import rank
from skimage import morphology
from skimage import segmentation
from skimage.exposure import rescale_intensity
from skimage.morphology import *
from scipy.ndimage import gaussian_filter
from skimage import img_as_float
from skimage.segmentation import flood, flood_fill
from skimage.filters import threshold_otsu
from skimage.segmentation import clear_border
from skimage.measure import label, regionprops
from skimage.morphology import closing, square
from skimage.color import label2rgb
from skimage.morphology import disk, ball
from skimage.filters import threshold_otsu, rank
from skimage.util import img_as_ubyte
from skimage.filters.rank import median
from skimage.filters import gaussian
from skimage.segmentation import active_contour
from skimage import io
from skimage import transform as tr
import skimage as skk

root = Tk()
root.title("IMAGE PROCESSING")
root.geometry("1366x720")
root.resizable(False, False)
Image_ = Image.open(os.path.abspath(os.getcwd()) +"\\no-image.jpg").convert('RGB')
img = Image_
img = ImageTk.PhotoImage(img)

temp = Image.open(os.path.abspath(os.getcwd()) +"\\no-image.jpg")
file_name = os.path.abspath(os.getcwd()) +"\\no-image.jpg"
panel = Label(root, image=img)
panel.image = img
panel.pack()
panel.place(relx=0.5, rely=0.5, anchor=CENTER) 

def openfn():
    filename = filedialog.askopenfilename(initialdir = os.getcwd() +"\\Samples",title='open',filetypes=[('all files', '.*'),
               ('image files', ('.png', '.jpg')),
           ])
    return filename

def open_img():
    x = openfn()
    global Image_
    Image_ = Image.open(x).convert('RGB')
    global file_name
    file_name = x
    img = ImageTk.PhotoImage(Image_)
    panel.configure(image=img)
    panel.image = img
    global temp
    temp = Image_
    
def savefile():
    filename = filedialog.asksaveasfile(mode='w', defaultextension=".png",filetypes=[('all files', '.*'),
               ('image files', ('.png')),
           ])
    if not filename:
        return
    global panel
    panel.image._PhotoImage__photo.write(filename.name)
    
def sobel():  
    global temp
    img_gray = rgb2gray(np.array(temp))
    edges = filters.sobel(img_gray)    
    low = 0.1
    high = 0.35
    lowt = (edges > low).astype(np.uint8)
    hight = (edges > high).astype(np.uint8)
    hyst = filters.apply_hysteresis_threshold(edges, low, high)
    img = Image.fromarray(lowt*255)
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img
    
def roberts():  
    global temp
    img_gray = rgb2gray(np.array(temp))
    edges = filters.roberts(img_gray)    
    img = Image.fromarray(edges*255)
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img
    
def gaussians(): 
    global temp
    img_gray = rgb2gray(np.array(temp))
    wimage = img_gray * window('hann', img_gray.shape)  # window image to improve FFT
    filtered_image = difference_of_gaussians(img_gray, 1, 12)  
    img = Image.fromarray(filtered_image*255)
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img

def laplace(): 
    global temp
    img_gray = rgb2gray(np.array(temp))
    wimage = img_gray * window('hann', img_gray.shape) 
    img = Image.fromarray(filters.laplace(wimage*255)*255)
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img
    
def hessian():
    global temp
    img_gray = rgb2gray(np.array(temp))
    wimage = img_gray * window('hann', img_gray.shape) 
    img = Image.fromarray(filters.hessian(wimage)*255)
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img

def unsharp_mask():  
    global temp
    img_gray = rgb2gray(np.array(temp))
    wimage = img_gray * window('hann', img_gray.shape) 
    img = Image.fromarray(filters.unsharp_mask(wimage)*255)
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img

def thresh_sauvola(): 
    global temp
    img_gray = rgb2gray(np.array(temp))
    wimage = img_gray * window('hann', img_gray.shape)     
    thresh_sauvola = filters.threshold_sauvola(wimage, window_size=25)
    img = Image.fromarray(thresh_sauvola*255)
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img

def sato():  
    global temp
    img_gray = rgb2gray(np.array(temp))
    wimage = img_gray * window('hann', img_gray.shape) 
    img = Image.fromarray(filters.sato(wimage)*255)
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img
    
def scharr():  
    global temp
    img_gray = rgb2gray(np.array(temp))
    wimage = img_gray * window('hann', img_gray.shape) 
    img = Image.fromarray(filters.scharr(wimage)*255)
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img
    
def meijering():  
    global temp
    img_gray = rgb2gray(np.array(temp))
    wimage = img_gray * window('hann', img_gray.shape) 
    img = Image.fromarray(filters.meijering(wimage)*255)
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img
    
def histogram_showing():
    img = np.array(temp)   
    %matplotlib qt
    hist(img.flatten(), 256, range=(0,256)); title('Histogram')
    
def histogram_matching():
    global temp
    img_gray = np.array(temp.convert('RGB'))
    reference = Image.open(openfn()).convert('RGB')    
    reference = np.array(reference)
    matched = match_histograms(img_gray, reference, multichannel=True)
    img = Image.fromarray(matched)
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img
       
def rescaling():
    global temp
    size = simpledialog.askstring("Input", "Value",
                                parent=root)
    img_  = np.array(temp)
    image_rescaled = rescale(img_ , float(size) , anti_aliasing=False, multichannel=True)
    img = Image.fromarray((image_rescaled* 255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img    
    
def resizing():
    global temp
    sizeX = simpledialog.askstring("Input", "X",
                                parent=root)
    sizeY = simpledialog.askstring("Input", "Y",
                                parent=root)
    img_gray = np.array(temp)
    image_resized = resize(img_gray, (float(sizeY), float(sizeX)),
                       anti_aliasing=True)
    img = Image.fromarray((image_resized * 255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img    
    
def shift_up10_left20(xy):
     return xy - np.array([-20, 10])[None, :]
    
def warp_coords_():
    global temp
    image = np.array(temp).astype(np.float32)
     
    coords = tr.warp_coords(shift_up10_left20, image.shape)
    warped_image = map_coordinates(image*255, coords)
    img = Image.fromarray((warped_image*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img  

def rotating():
    global temp
    size = simpledialog.askstring("Input", "Value",
                                parent=root)
    img_ = np.array(temp)
    image_rescaled = rotate(img_, float(size) , resize=True)
    img = Image.fromarray((image_rescaled*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img    
    
def swirl_():  
    global temp
    img =  swirl(np.array(temp), rotation=0, strength=10, radius=120)       
    img = Image.fromarray((img*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img

def intensity():  
    global temp
    one = simpledialog.askstring("Input", "one",
                                parent=root)
    two = simpledialog.askstring("Input", "two",
                                parent=root)
    img =  rescale_intensity(rgb2gray(np.array(temp))*255, out_range=(one, two))       
    img = Image.fromarray(img)
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img
    
def erosion_():
    global temp
    image = np.array(temp).astype(np.float32)

    chull = erosion(image)
    img = Image.fromarray((chull*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img  

def binary_opening_():
    global temp
    image = np.array(temp).astype(np.float32)

    chull = binary_opening(image)
    img = Image.fromarray((chull*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img  
    
def binary_dilation_():
    global temp
    image = np.array(temp).astype(np.float32)
    chull = binary_dilation(image)
    img = Image.fromarray((chull*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img  

def convex_hull_image_():
    global temp
    image = np.array(temp).astype(np.float32)
    chull = convex_hull_image(image)
    img = Image.fromarray((chull*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img 
    
def diameter_closing_():
    global temp
    image = np.array(temp).astype(np.float32)
    chull = diameter_closing(image)
    img = Image.fromarray((chull*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img 

def dilation_():
    global temp
    image = np.array(temp).astype(np.float32)
    chull = dilation(image)
    img = Image.fromarray((chull*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img 

def flood_():
    global temp
    image = np.array(temp).astype(np.float32)
    chull = flood(image)
    img = Image.fromarray((chull*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img 

def local_maxima_():
    global temp
    image = np.array(temp).astype(np.float32)
    chull = local_maxima(image)
    img = Image.fromarray((chull*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img  
    
def local_minima_():
    global temp
    image = np.array(temp).astype(np.float32)
    chull = binary_dilation(image)
    img = Image.fromarray((chull*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img 

def watershed_():
    global temp
    image = np.array(temp).astype(np.float32)
    chull = watershed(image)
    img = Image.fromarray((chull*255).astype(np.uint8)).convert("RGB")
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img  

def active_contour_():
    img = color.rgb2gray(np.array(temp))   
    %matplotlib qt

    s = np.linspace(0, 2*np.pi, 400)
    x = 220 + 100*np.cos(s)
    y = 100 + 100*np.sin(s)
    init = np.array([x, y]).T

    snake = active_contour(gaussian(img, 3),
                           init, alpha=0.015, beta=10, gamma=0.001)

    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_subplot(111)
    plt.gray()
    ax.imshow(img)
    ax.plot(init[:, 0], init[:, 1], '--r', lw=3)
    ax.plot(snake[:, 0], snake[:, 1], '-b', lw=3)
    ax.set_xticks([]), ax.set_yticks([])
    ax.axis([0, img.shape[1], img.shape[0], 0])
    plt.show()

def face_detection_():
    
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    cap = cv2.VideoCapture(0)

    while True:
        _, img = cap.read()
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.1, 4)
        for (x, y, w, h) in faces:
            cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
        cv2.imshow('Stop if escape key is pressed', img)
        k = cv2.waitKey(30) & 0xff
        if k==27:
            break
    cap.release()

def put_glasses(glass, fc, x, y, w, h):
    width = w
    height = h

    hat_width = width + 1
    hat_height = int(0.50 * height) + 1

    glass = cv2.resize(glass, (hat_width, hat_height))

    for i in range(hat_height):
        for j in range(hat_width):
            for k in range(3):
                if glass[i][j][k] < 235:
                    fc[y + i - int(-0.20 * height)][x + j][k] = glass[i][j][k]
    return fc


def snap_():
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    hat = io.imread(os.path.abspath(os.getcwd()) +"\\glasses.png")
    img= io.imread(file_name)
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    fl=face_cascade.detectMultiScale(gray,1.09,7)
    for (x, y, w, h) in fl:
        frame = put_glasses(hat, img, x, y, w, h)
    reference = Image.open(os.path.abspath(os.getcwd()) +"\\Samples\\color.png").convert('RGB')    
    reference = np.array(reference)
    matched = match_histograms(frame, reference, multichannel=True)
    img = Image.fromarray(matched)
    global temp
    temp = img
    img = ImageTk.PhotoImage(img)
    panel.configure(image=img)
    panel.image = img  

menubar = Menu(root)
filemenu = Menu(menubar, tearoff=0)
filemenu.add_command(label="Open Image", command=open_img)
filemenu.add_command(label="Save Image", command=savefile)

menubar.add_cascade(label="File", menu=filemenu)

Enhancement = Menu(menubar, tearoff=0)
Enhancement.add_command(label="Sobel", command=sobel)
Enhancement.add_command(label="Roberts", command=roberts)
Enhancement.add_command(label="Gaussians", command=gaussians)
Enhancement.add_command(label="Laplace", command=laplace)
Enhancement.add_command(label="Hessian", command=hessian)
Enhancement.add_command(label="Unsharp Mask", command=unsharp_mask)
Enhancement.add_command(label="Threshold Sauvola", command=thresh_sauvola)
Enhancement.add_command(label="Sato", command=sato)
Enhancement.add_command(label="Scharr", command=scharr)
Enhancement.add_command(label="Meijering", command=meijering)

menubar.add_cascade(label="Image Enhancement", menu=Enhancement)

histogram = Menu(menubar, tearoff=0)
histogram.add_command(label="Histogram Showing", command=histogram_showing)
histogram.add_command(label="Histogram Matching", command=histogram_matching)

menubar.add_cascade(label="Histogram", menu=histogram)

transform = Menu(menubar, tearoff=0)
transform.add_command(label="Rescaling", command=rescaling)
transform.add_command(label="Resizing", command=resizing)
transform.add_command(label="Warp Coords", command=warp_coords_)
transform.add_command(label="Rotating", command=rotating)
transform.add_command(label="Swirl", command=swirl_)

menubar.add_cascade(label="Transform", menu=transform)

intensity_ = Menu(menubar, tearoff=0)
intensity_.add_command(label="Rescaling Intensity", command=intensity)

menubar.add_cascade(label="Intensity", menu=intensity_)

morphology_ = Menu(menubar, tearoff=0)
morphology_.add_command(label="Erosion", command=erosion_)
morphology_.add_command(label="Binary Opening", command=binary_opening_)
morphology_.add_command(label="Binary Dilation", command=binary_dilation_)
morphology_.add_command(label="Convex Hull Image", command=convex_hull_image_)
morphology_.add_command(label="Diameter Closing", command=diameter_closing_)
morphology_.add_command(label="Dilation", command=dilation_)
morphology_.add_command(label="Flood", command=flood_)
morphology_.add_command(label="Local Maxima", command=local_maxima_)
morphology_.add_command(label="Local Minima", command=local_minima_)
morphology_.add_command(label="Watershed", command=watershed_)


menubar.add_cascade(label="Morphology", menu=morphology_)

active_ = Menu(menubar, tearoff=0)
active_.add_command(label="Active Contour", command=active_contour_)

menubar.add_cascade(label="Active Contour", menu=active_)

facemenu = Menu(menubar, tearoff=0)
facemenu.add_command(label="Face Detection", command=face_detection_)

menubar.add_cascade(label="OpenCV Video", menu=facemenu)

snap = Menu(menubar, tearoff=0)
snap.add_command(label="Snap", command=snap_)

menubar.add_cascade(label="Snap Filter", menu=snap)



root.config(menu=menubar)
root.mainloop()
```
